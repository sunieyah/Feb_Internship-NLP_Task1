## Task 1: Conceptual Understanding

### 1. Difference between "Love" and "love" in NLP  
In Natural Language Processing (NLP), "Love" and "love" are treated as different tokens because most NLP systems are case-sensitive by default. For instance, "Love" may appear at the beginning of a sentence or be used as a proper noun, whereas "love" is generally used as a common word. To avoid such inconsistencies and duplication, text is usually converted into lowercase during preprocessing.

---

### 2. What happens if stopwords are not removed?  
If stopwords are not removed, the dataset will contain many frequently used words such as "is", "the", and "and" that do not carry significant meaning. This increases noise in the data, makes processing slower, and can negatively impact the performance of machine learning models since these words do not contribute much to understanding the actual context.

---

### 3. Two scenarios where removing stopwords can be harmful  

- **Sentiment Analysis:**  
  Words like "not" are crucial for understanding sentiment. For example, "not good" has the opposite meaning of "good". Removing such words can lead to incorrect interpretation of sentiment.

- **Question Answering Systems:**  
  Words like "who", "what", "where", and "when" are essential to understand the intent of a question. Removing them can make it difficult for the system to interpret and respond correctly.

---

### 4. Difference between stemming and lemmatization  

- **Stemming:**  
  Stemming is a basic technique that reduces words to their root form by removing suffixes. For example, "running" becomes "run". However, it may sometimes produce incorrect or incomplete words.

- **Lemmatization:**  
  Lemmatization is a more advanced technique that converts words into their correct base or dictionary form using linguistic rules. For example, "better" becomes "good". It is more accurate than stemming but slightly slower.

## TASK 2: ADVANCED PREPROCESS FUNCTION

```python
import re

def normalize_repeated_chars(word):

    # Step 1: reduce 3+ repeats → 2
    word = re.sub(r'(.)\1{2,}', r'\1\1', word)

    # Step 2: if still exaggerated (like 'soo'), reduce to single
    if len(set(word)) == 1:  # handles 'aaa'
        word = word[0]

    return word


def preprocess_text(text):
    if not text or not isinstance(text, str):
        return []

    # Lowercase
    text = text.lower()

    # Remove URLs & emails
    text = re.sub(r'http\S+|www\S+|\S+@\S+', '', text)

    # Remove numbers
    text = re.sub(r'\d+', '', text)

    # Remove special characters (keep spaces)
    text = re.sub(r'[^a-z\s]', '', text)

    # Remove extra spaces
    text = re.sub(r'\s+', ' ', text).strip()

    # Tokenization
    tokens = text.split()

    # Normalize repeated characters word-by-word
    tokens = [normalize_repeated_chars(word) for word in tokens]

    # Remove short tokens except 'no', 'not'
    tokens = [word for word in tokens if len(word) > 2 or word in ['no', 'not']]

    return tokens


    # Example test cases

examples = [
    "I have 2 dogs",
    "This   is   good",
    "sooo goooood!!!",
    "WOW!!! This is GREAT!!!",
    "Visit http://example.com now"
]

for text in examples:
    print("Input :", text)
    print("Output:", " ".join(preprocess_text(text)))
    print("-" * 50)
```

## TASK 3: Stress Testing

```python
# Test inputs (given in assignment)

test_sentences = [
    "Get 100% FREE access now!!!",
    "I absolutely looooved this product 😍😍",
    "Worst service ever... 0/10",
    "Call me at 9876543210",
    "This is THE best course!!!",
    "Visit https://openai.com now!",
    "Nooooo this is baaad!!!",
    "OK OK OK I got it",
    "Win $$$ now!!! Limited offer!!!",
    "I am not happy with this"
]

# Display results in required format

for i, sentence in enumerate(test_sentences, 1):
    tokens = preprocess_text(sentence)
    cleaned_sentence = " ".join(tokens)

    print(f"Sentence {i}")
    print("Original Text   :", sentence)
    print("Cleaned Tokens  :", tokens)
    print("Cleaned Sentence:", cleaned_sentence)
    print("=" * 60)
```

## TASK 4: Token Analytics

```python
print("Token Analytics:\n")

analytics_results = []

for i, sentence in enumerate(test_sentences, 1):
    tokens = preprocess_text(sentence)

    total_tokens = len(tokens)
    unique_tokens = len(set(tokens))

    if total_tokens > 0:
        avg_length = sum(len(word) for word in tokens) / total_tokens
    else:
        avg_length = 0

    print(f"Sentence {i}")
    print("Tokens          :", tokens)
    print("Total Tokens    :", total_tokens)
    print("Unique Tokens   :", unique_tokens)
    print("Avg Token Length:", round(avg_length, 2))
    print("=" * 60)

    analytics_results.append((sentence, tokens, total_tokens, unique_tokens, avg_length))
```

## TASK 5: Frequency Analysis

```python
from collections import Counter

# Combine all tokens from previous results
all_tokens = []

for sentence in test_sentences:
    tokens = preprocess_text(sentence)
    all_tokens.extend(tokens)

# Create frequency counter
word_freq = Counter(all_tokens)

# Top 10 most frequent words
print("Top 10 Most Frequent Words:")
for word, count in word_freq.most_common(10):
    print(f"{word} : {count}")

print("\n" + "="*50)

# Top 5 least frequent words
print("Top 5 Least Frequent Words:")
least_common = word_freq.most_common()[::-1][:5]

for word, count in least_common:
    print(f"{word} : {count}")
```

## TASK 6: Full Pipeline

```python
def full_pipeline(text_list):
    all_tokens = []
    clean_sentences = []

    for text in text_list:
        tokens = preprocess_text(text)

        all_tokens.extend(tokens)
        clean_sentences.append(" ".join(tokens))

    return {
        "tokens": all_tokens,
        "clean_sentences": clean_sentences
    }

# Test the pipeline
pipeline_output = full_pipeline(test_sentences)

print("Clean Sentences:\n")
for sentence in pipeline_output["clean_sentences"]:
    print(sentence)

print("\n" + "="*50)
print("All Tokens:\n")
print(pipeline_output["tokens"])
```
